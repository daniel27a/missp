```
"""
The Simplified Whiskas Model Python Formulation for the PuLP Modeller

"""

from pulp import *

prob = LpProblem("The Whiskas' Cat Food Problem",LpMinimize)

# The 2 variables Beef and Chicken are created with a lower limit of zero
x1=LpVariable("ChickenPercent",0,)
x2=LpVariable("BeefPercent",0)
x3=LpVariable("SardinesPercent",0)

# The objective function is added to 'prob' first
prob += 0.013*x1 + 0.008*x2 + 0.01*x3, "Total Cost of Ingredients per can"

# The five constraints are entered
prob += x1 + x2 + x3 == 100, "PercentagesSum"
prob += 0.100*x1 + 0.200*x2 + 0.25*x3 >= 8.0, "ProteinRequirement"
prob += 0.080*x1 + 0.100*x2 + 0.11*x3 >= 6.0, "FatRequirement"
prob += 0.001*x1 + 0.005*x2 + 0*x3 <= 2.0, "FibreRequirement"
prob += 0.002*x1 + 0.005*x2 + 0.00505*x3 <= 0.4, "SaltRequirement"
prob += 0.003*x1 + 0.026*x2 + 0.089*x3 >= 0.8, "VitaminBreq" 

prob.writeLP("WhiskasModel.lp")

prob.solve()
print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)

produkty_procentowo = ""

for v in prob.variables():
    produkty_procentowo += v.name + " = " + str(v.varValue)+ "%, " 

produkty_procentowo = produkty_procentowo[:-2]

print("Puszka produktu będzie kosztować ", value(prob.objective), "$ i będzie miał w składzie:", produkty_procentowo,".")

print("-"*30)
```

```
"""
Windows problem with sliders

"""

# W przypadku niezmieniancyh danych problem rozwiązuje się dla danych podancyh na zajęciach.

from pulp import *
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

#slidery
zysk_z_aluminiowych_slider = IntSlider (min=0,max = 10000, value= 3000, description="zysk_z_aluminiowych", style = style)
zysk_z_drewnianych_slider  = IntSlider (min=0,max = 10000, value= 5000, description="zysk_z_drewnianych",  style = style)
f1_caly_czas_slider        = IntSlider (min=0,max = 20,    value= 4,    description="f1_caly_czas",        style = style)
f2_caly_czas_slider        = IntSlider (min=0,max = 20,    value= 12,   description="f2_caly_czas",        style = style)
f3_caly_czas_slider        = IntSlider (min=0,max = 20,    value= 18,   description="f3_caly_czas",        style = style)
f1_czas_na_alum_slider     = IntSlider (min=0,max = 10,    value= 1,    description="f1_czas_na_alum",     style = style )
f2_czas_na_alum_slider     = IntSlider (min=0,max = 10,    value= 0,    description="f2_czas_na_alum",     style = style )
f3_czas_na_alum_slider     = IntSlider (min=0,max = 10,    value= 3,    description="f3_czas_na_alum",     style = style )
f1_czas_na_drewno_slider   = IntSlider (min=0,max = 10,    value= 0,    description="f1_czas_na_drewno",   style = style )
f2_czas_na_drewno_slider   = IntSlider (min=0,max = 10,    value= 2,    description="f2_czas_na_drewno",   style = style )
f3_czas_na_drewno_slider   = IntSlider (min=0,max = 10,    value= 2,    description="f3_czas_na_drewno",   style = style )

#definicja wartosci domyslnych
def produckja_okien( 
    zysk_z_aluminiowych= 3000,
    zysk_z_drewnianych = 5000,
    f1_caly_czas = 4,
    f2_caly_czas = 12,
    f3_caly_czas = 18,
    f1_czas_na_alum = 1,
    f2_czas_na_alum = 0,
    f3_czas_na_alum = 3,
    f1_czas_na_drewno = 0,
    f2_czas_na_drewno = 2,
    f3_czas_na_drewno = 2):

# The Main problem
    prob = LpProblem("Windows Problem",LpMaximize)


# The 2 variables, alum windows and wood windows
    x1 = LpVariable("Okna aluminiowe",0,None,LpInteger)       
    x2 = LpVariable("Okna drewniane",0,None,LpInteger)


# Profit maximalization 
    prob += zysk_z_aluminiowych*x1 + zysk_z_drewnianych*x2, "Total profit from windows sales"


# Limitations
    prob += f1_czas_na_alum*x1 + f1_czas_na_drewno*x2 <= f1_caly_czas, "Factory 1"
    prob += f2_czas_na_alum*x1 + f2_czas_na_drewno*x2 <= f2_caly_czas, "Factory 2"
    prob += f3_czas_na_alum*x1 + f3_czas_na_drewno*x2 <= f3_caly_czas, "Factory 3"

# Solution
    prob.solve()
    print("Status:", LpStatus[prob.status],)

    print("Wyprodukowano:")
    for v in prob.variables():
      print(v.name, ":", int(v.varValue))
    print("Profit:",value(prob.objective),"zł")

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))

    print("-"*50)

# sliders handlers
interact(produckja_okien,
    zysk_z_aluminiowych= zysk_z_aluminiowych_slider,
    zysk_z_drewnianych = zysk_z_drewnianych_slider,
    f1_caly_czas = f1_caly_czas_slider,
    f2_caly_czas = f2_caly_czas_slider,
    f3_caly_czas = f3_caly_czas_slider,
    f1_czas_na_alum = f1_czas_na_alum_slider,
    f2_czas_na_alum = f2_czas_na_alum_slider,
    f3_czas_na_alum = f3_czas_na_alum_slider,
    f1_czas_na_drewno = f1_czas_na_drewno_slider,
    f2_czas_na_drewno = f2_czas_na_drewno_slider,
    f3_czas_na_drewno = f3_czas_na_drewno_slider)
```

```
"""
Minimalizacja mocy rozproszonej - mostek rezystorowy

https://www.youtube.com/watch?v=yZuVfekIXJ0

"""

import numpy as np
from pulp import *

prob = LpProblem("Minimalizacja mocy rozproszonej",LpMinimize)

# Variables (name, Lo, Hi, Type)

i1=LpVariable("I1", 0.003, 0.005, LpContinuous)
i2=LpVariable("I2", 0.001, 0.003, LpContinuous)
i3=LpVariable("I3", 0.001, 0.003, LpContinuous)
i4=LpVariable("I4", 0.001, 0.003, LpContinuous)
i5=LpVariable("I5", 0.003, 0.005, LpContinuous)
 

# Constants
U = (6.0, 10.0, 4.0, 7.0, 3.0)


# Main problem
prob += U[0]*i1 + U[1]*i2 + U[2]*i3 + U[3]*i4 + U[4]*i5, "Total Power"


# Constraints
prob += i1-i3 == i4, "KI1" 
prob += i2+i3 == i5, "KI2"

prob.writeLP("ResistorBridge.lp")

prob.solve()
print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)

print()

print("Straty:", value(prob.objective), "W\n")  

print("-"*30,"\n")

for v, k in zip(U, prob.variables()):
  print("R",U.index(v)+1,"=",(v/k.varValue/1000),"k ohm") 
```

```
"""
Maksymalizacja prądu - dwójnik rezystancyjny

https://www.youtube.com/watch?v=buPWuzar384

"""

import numpy as np
from pulp import *

prob = LpProblem("Maksymalizacja prądu w dwójniku rezystorowym",LpMaximize)

# Variables (name, Lo, Hi, Type)

i1=LpVariable("I1", 0, 2, LpContinuous)
i2=LpVariable("I2", 0, 3, LpContinuous)
i3=LpVariable("I3", 0, 4, LpContinuous)
i4=LpVariable("I4", 0, 2, LpContinuous)
i5=LpVariable("I5", 0, 2, LpContinuous)
 

# Constants
R = (8.0, 6.0, 4.0, 10.0, 8.0)

# Main problem
prob += i1+i2+i3+i4+i5, "max current"

# Constraints
prob += i1+i2 == i3, "1" 
prob += i3-i4 == i5, "2"
prob += i1*R[0] - i2*R[1] == 0, "3"
prob += i4*R[3] - i5*R[4] == 0, "4"

prob.writeLP("dwojnik.lp")

prob.solve()
print("Status:", LpStatus[prob.status])

Imax = 0
for v in prob.variables():
    print(v.name, "=", v.varValue, "A")
    if (v.varValue >= Imax):
      Imax = v.varValue 
    
print()
print("Suma prądów:", value(prob.objective), "A\n") 
print("Imax = ", Imax, "A\n") 
```

```
"""
Optymalizacja pracowników

https://www.youtube.com/watch?v=Our22Z6-FDg

"""


from pulp import *

prob = LpProblem("Optymalizacja_pracownikow",LpMinimize)

### Variables (name, Lo, Hi, Type) Ania Stefan Hektor Olaf Lidia Piotr
A1 = LpVariable("Ania_pon", 0, 1, LpInteger)
A2 = LpVariable("Ania_wt", 0, 1, LpInteger)
A3 = LpVariable("Ania_sr", 0, 1, LpInteger)
A4 = LpVariable("Ania_czw", 0, 1, LpInteger)
A5 = LpVariable("Ania_pt", 0, 1, LpInteger)
S1 = LpVariable("Stefan_pon", 0, 1, LpInteger)
S2 = LpVariable("Stefan_wt", 0, 1, LpInteger)
S3 = LpVariable("Stefan_sr", 0, 1, LpInteger)
S4 = LpVariable("Stefan_czw", 0, 1, LpInteger)
S5 = LpVariable("Stefan_pt", 0, 1, LpInteger)
H1 = LpVariable("Hektor_pon", 0, 1, LpInteger)
H2 = LpVariable("Hektor_wt", 0, 1, LpInteger)
H3 = LpVariable("Hektor_sr", 0, 1, LpInteger)
H4 = LpVariable("Hektor_czw", 0, 1, LpInteger)
H5 = LpVariable("Hektor_pt", 0, 1, LpInteger)
O1 = LpVariable("Olaf_pon", 0, 1, LpInteger)
O2 = LpVariable("Olaf_wt", 0, 1, LpInteger)
O3 = LpVariable("Olaf_sr", 0, 1, LpInteger)
O4 = LpVariable("Olaf_czw", 0, 1, LpInteger)
O5 = LpVariable("Olaf_pt", 0, 1, LpInteger)
L1 = LpVariable("Lidia_pon", 0, 1, LpInteger)
L2 = LpVariable("Lidia_wt", 0, 1, LpInteger)
L3 = LpVariable("Lidia_sr", 0, 1, LpInteger)
L4 = LpVariable("Lidia_czw", 0, 1, LpInteger)
L5 = LpVariable("Lidia_pt", 0, 1, LpInteger)
P1 = LpVariable("Piotr_pon", 0, 1, LpInteger)
P2 = LpVariable("Piotr_wt", 0, 1, LpInteger)
P3 = LpVariable("Piotr_sr", 0, 1, LpInteger)
P4 = LpVariable("Piotr_czw", 0, 1, LpInteger)
P5 = LpVariable("Piotr_pt", 0, 1, LpInteger)


### Constants
Ac = 100
Sc = 50
Hc = 60
Oc = 40
Lc = 110
Pc = 70

### Main problem
prob += (A1+A2+A3+A4+A5) * Ac + (S1+S2+S3+S4+S5) * Sc + (H1+H2+H3+H4+H5) * Hc + (O1+O2+O3+O4+O5) * Oc + (L1+L2+L3+L4+L5) * Lc + (P1+P2+P3+P4+P5) * Pc, "min_cost"

### Constraints
# ilość dniówek
prob += A1+A2+A3+A4+A5 <= 3, "1" 
prob += S1+S2+S3+S4+S5 <= 3, "2" 
prob += H1+H2+H3+H4+H5 <= 3, "3" 
prob += O1+O2+O3+O4+O5 <= 3, "4"
prob += L1+L2+L3+L4+L5 <= 3, "5"
prob += P1+P2+P3+P4+P5 <= 3, "6"

# kiedy pracownicy mogą pracować
prob += S2+S3+S4+S5 == 0 , "7"
prob += H4+H5 == 0, "8"
prob += L1+L2+L3 == 0, "9"
prob += P4+P5 == 0, "10"

#ilość ludzi na sklepie w dany dzień
prob += A1+S1+H1+O1+L1+P1 == 2, "Pon"
prob += A2+S2+H2+O2+L2+P2 == 1, "Wt"
prob += A3+S3+H3+O3+L3+P3 == 1, "Sr"
prob += A4+S4+H4+O4+L4+P4 == 1, "Czw"
prob += A5+S5+H5+O5+L5+P5 == 3, "Pt"


prob.writeLP("week.lp")

prob.solve()
print("Status:", LpStatus[prob.status])

print("Calkowity koszt:", value(prob.objective), "zl\n") 

for v in prob.variables():
    print(v.name, "=", v.varValue)
```
